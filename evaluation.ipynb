```python
from evaluate import *
```

```python
config = EvaluationConfig(
    scene_name="hyfluid",
    pretrained_ckpt="history/ckpt_hyfluid_0415071408_100000.tar",
    target_device=torch.device("cuda" if torch.cuda.is_available() else "cpu"),
    target_dtype=torch.float32,
)
model_hyfluid = EvaluationSpatialState(config, 128, 192, 128)
density_flat = model_hyfluid.plot_density_distribution_histogram(100, bins=10)
```

```python
# 计算最大值和最小值
vmax = np.max(density_flat)
vmin = np.min(density_flat)

# 生成从 vmax 到 vmin 的 11 个边界点（共10个区间）
bins = np.linspace(vmin, vmax, 11)

# 使用 np.histogram 统计每一段的个数
counts, edges = np.histogram(density_flat, bins=bins)

# 打印每一段信息
for i in range(len(counts)):
    print(f"区间 {edges[i]:.2f} ~ {edges[i+1]:.2f}：{counts[i]} 个元素")
```

```python
median = np.quantile(density_flat, 0.5)
percentile_10 = np.percentile(density_flat, 10)
percentile_90 = np.percentile(density_flat, 90)
print(f"中位数: {median}, 10th 百分位数: {percentile_10}, 90th 百分位数: {percentile_90}")
```
